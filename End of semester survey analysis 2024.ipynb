```r
# Drop NaNs from Q21 (Python)
df_Spring = df_Spring[df_Spring["Q21"].notna()]
df_fall = df_fall[df_fall["Q21"].notna()]

# Use consistent variables
programming_questions = ['Q8', 'Q9', 'Q10', 'Q11', 'Q12', 'Q13', 'Q14', 'Q15', 'Q16', 'Q17',
                         'Q18', 'Q19', 'Q20', 'Q21', 'Q22', 'Q23', 'Q24']

programming_spring_responses = df_Spring[programming_questions]
programming_fall_responses = df_fall[programming_questions]

# Count responses
programming_spring_counts = programming_spring_responses.apply(pd.Series.value_counts)
programming_fall_counts = programming_fall_responses.apply(pd.Series.value_counts)

# Convert to percentages
programming_spring_percentage = programming_spring_counts.div(programming_spring_counts.sum(axis=0), axis=1) * 100
programming_fall_percentage = programming_fall_counts.div(programming_fall_counts.sum(axis=0), axis=1) * 100

# Fill NaNs (to prevent issues when subtracting)
programming_spring_percentage.fillna(0, inplace=True)
programming_fall_percentage.fillna(0, inplace=True)

# Calculate % change
programming_pct_change = programming_fall_percentage - programming_spring_percentage

# Show result
programming_pct_change.round(2)

```

```r
# Loop through programming questions and run chi-square test
for (q in programming_qs) {
  cat("\n\n---", q, "---\n")
  print(table(df_all$Semester, df_all[[q]]))
  
  # Chi-square test only if more than one unique value
  if (length(na.omit(unique(df_all[[q]]))) > 1) {
    test <- chisq.test(table(df_all$Semester, df_all[[q]]))
    print(test)
  } else {
    cat("Not enough unique values to test.\n")
  }
}
```

```r
# Interpretation:
# For all programming skill questions (Q8 to Q24), Chi-square tests were used to check
# if the distribution of self-reported skill levels significantly differed between Spring and Fall 2024.
# 
# Across all questions, p-values were greater than 0.05, indicating that the differences
# in response distributions between semesters were not statistically significant.
# 
# In simpler terms: Although there are slight shifts in the number of students selecting 
# each skill level (e.g., more "No experience" or fewer "Strong ability" in some questions), 
# these shifts are not large enough to conclude that a real semester-based change occurred.
# 
# Note: Some warnings ("Chi-squared approximation may be incorrect") occurred due to small 
# response counts in certain categories, which can affect the test's reliability.
```

```r
library(MASS)

# Store question codes
programming_qs <- paste0("Q", 8:24)

# Loop through each programming question
for (q in programming_qs) {
  cat("\n\n---", q, "---\n")
  
  # Subset: remove NA
  df_q <- na.omit(df_all[, c("Semester", q)])
  
  # Convert response to ordered factor
  df_q[[q]] <- factor(df_q[[q]], ordered = TRUE,
                      levels = c("No experience",
                                 "Don't know now, but interested in learning",
                                 "Could get by with help from others",
                                 "Some ability",
                                 "Strong Ability"))
  
  # Fit model
  model <- polr(as.formula(paste(q, "~ Semester")), data = df_q, Hess = TRUE)
  
  # Get p-values
  coef_table <- coef(summary(model))
  p_vals <- pnorm(abs(coef_table[, "t value"]), lower.tail = FALSE) * 2
  coef_table <- cbind(coef_table, "p value" = round(p_vals, 4))
  
  print(coef_table)
}
```

```r
# Interpretation of ordinal regression results (Q8 to Q21)
# ---------------------------------------------------------

# Q8: No significant difference between Fall and Spring (p = 0.7073).
#     Semester does not influence ability levels. Thresholds are valid and highly significant.

# Q9: Semester not significant (p = 0.6347). Clear ability progression, strong threshold support.

# Q10: No semester effect (p = 0.6996). Thresholds highly significant; strong scale structure.

# Q11: Semester not a significant predictor (p = 0.9403). Very robust threshold structure.

# Q12: Semester not significant (p = 0.2677). One lower threshold not significant, others are strong.

# Q13: Semester shows a small but significant effect (p = 0.0401).
#      Fall and Spring differ slightly in skill perception. All thresholds are strong — reportable.

# Q14: Semester not significant (p = 0.1162), but borderline enough to explore in follow-ups.
#      Scale thresholds are highly significant.

# Q15: Semester is not significant (p = 0.1998). All ability cutoffs are statistically strong.

# Q16: Semester again not significant (p = 0.3198). Strong and meaningful thresholds.

# Q17: Semester not significant (p = 0.4214). Threshold model fits well.

# Q18: Semester not significant (p = 0.1195), but effect is larger than earlier questions.
#      All transitions between ability levels are solidly significant.

# Q19: Semester not significant (p = 0.6424). Thresholds hold up very well.

# Q20: Semester not significant (p = 0.1859). Note: one ability transition is not significant (t = 0.88).

# Q21: ERROR — Ordinal regression failed due to only 3 response levels (e.g., Python: Some ability / Strong / Help from others).
#      Consider a multinomial logistic regression or alternate test.
```

```r
# Load required package
library(nnet)

# Prepare data for Q21
df_q21 <- df_all[, c("Semester", "Q21")]
df_q21 <- na.omit(df_q21)  # drop NAs

# Make sure Q21 is a factor (multinomial needs this)
df_q21$Q21 <- factor(df_q21$Q21)

# Fit multinomial logistic regression
model_q21 <- multinom(Q21 ~ Semester, data = df_q21)

# Summary of model
summary(model_q21)

# Get p-values using z-test
z_vals <- summary(model_q21)$coefficients / summary(model_q21)$standard.errors
p_vals <- 2 * (1 - pnorm(abs(z_vals)))
p_vals
```

```r
# Define ordered levels
ordered_levels <- c(
  "No experience",
  "Don't know now, but interested in learning",
  "Could get by with help from others",
  "Some ability",
  "Strong Ability"
)

# Apply ordered factor to each programming question
for (q in programming_qs) {
  if (q %in% colnames(df_all)) {
    df_all[[q]] <- factor(trimws(df_all[[q]]), levels = ordered_levels, ordered = TRUE)
  }
}
```

```r
# Perform Kruskal-Wallis test for each programming question (Q8-Q24)
for (q in programming_qs) {
  if (q %in% colnames(df_all)) {
    # Run Kruskal-Wallis test
    test_result <- kruskal.test(df_all[[q]] ~ df_all$Semester)
    
    # Print the test result for each question
    cat("\n---", q, "---\n")
    print(test_result)
  }
}
```
